```python
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [12, 6]
```

```python
from dacbench.benchmarks import LubyBenchmark
from dacbench.wrappers import EpisodeTimeWrapper, PerformanceTrackingWrapper, ActionFrequencyWrapper, StateTrackingWrapper, InstanceSamplingWrapper, RewardNoiseWrapper
from examples.example_utils import q_learning
```

```python
# Get replica environment
bench = LubyBenchmark()
env = bench.get_benchmark(L=8, fuzziness=1)
```

```python
# Get base env
bench.config.seed = 0
env = bench.get_benchmark_env()
```

```python
Q, stats = q_learning(env, num_episodes=5)
print(f"Rewards: {stats[1].episode_rewards}")
```

```python
env = bench.get_benchmark_env()
env = RewardNoiseWrapper(env)
Q, stats = q_learning(env, num_episodes=5)
print(f"Rewards: {stats[1].episode_rewards}")
```

```python
env = bench.get_benchmark_env()
env = RewardNoiseWrapper(env, noise_dist="exponential", dist_args=[0.3])
Q, stats = q_learning(env, num_episodes=5)
print(f"Rewards: {stats[1].episode_rewards}")
```

```python
def dummy_noise():
    return 1
env = bench.get_benchmark_env()
env = RewardNoiseWrapper(env, noise_function=dummy_noise)
Q, stats = q_learning(env, num_episodes=5)
print(f"Rewards: {stats[1].episode_rewards}")
```

```python
env = bench.get_benchmark_env()
print(env.instance_set)
```

```python
new_instance_set = [[1, 2], [3, 4]]
env.instance_set = new_instance_set
print(env.instance_set)
```

```python
bench.config.instance_set_path = "../instance_sets/luby_train.csv"
bench.read_instance_set()
env = bench.get_benchmark_env()
print(env.instance_set)
```

```python
def sample_luby():
    return [np.random.uniform(-5, 5), np.random.uniform(-1, 1)]
env = bench.get_benchmark_env()
env = InstanceSamplingWrapper(env, sampling_function=sample_luby)
print(env.instance_set)
env.reset()
print(env.instance_set)
```

```python
env = bench.get_benchmark_env()
env = InstanceSamplingWrapper(env, instances=env.instance_set)
print(env.instance_set)
env.reset()
print(env.instance_set)
```

```python
env = bench.get_benchmark_env()
env = InstanceSamplingWrapper(env, instances=env.instance_set)
env = RewardNoiseWrapper(env, noise_dist="normal", dist_args=[0, 0.5])
env = ActionFrequencyWrapper(env)
env = StateTrackingWrapper(env)
env = EpisodeTimeWrapper(env, 5)
env = PerformanceTrackingWrapper(env)
```

```python
Q, stats = q_learning(env, num_episodes=20, verbose=True)
print(f"Rewards: {stats[1].episode_rewards}")
print(sum(stats[1].episode_lengths))
```

```python
overall_performance, instance_performance = env.get_performance()
```

```python
plt.plot(np.arange(len(overall_performance)/2), overall_performance[1::2])
plt.xlabel("Step")
plt.ylabel("Reward")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.title("Mean Performance per Instance")
plt.ylabel("Mean reward")
plt.xlabel("Instance")
ax = plt.subplot(111)
for k, i in zip(instance_performance.keys(), np.arange(len(instance_performance.keys()))):
    ax.bar(str(i), np.mean(instance_performance[k]))
plt.show()
```
